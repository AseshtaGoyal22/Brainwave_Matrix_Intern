```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/Top 50 Fast-Food Chains in USA.csv")

# Set Seaborn style
sns.set(style="whitegrid")

# Top 10 Chains by Total U.S. Systemwide Sales
top_sales = df.sort_values(by="U.S. Systemwide Sales (Millions - U.S Dollars)", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_sales, x="U.S. Systemwide Sales (Millions - U.S Dollars)", y="Fast-Food Chains", hue="Fast-Food Chains", dodge=False)
plt.title("Top 10 Fast Food Chains by Total U.S. Sales")
plt.xlabel("Sales (in millions USD)")
plt.ylabel("Fast Food chain")
plt.tight_layout()
plt.show()

# Top 10 by Average Sales per Unit
top_avg_sales = df.sort_values(by="Average Sales per Unit (Thousands - U.S Dollars)", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_avg_sales, x="Average Sales per Unit (Thousands - U.S Dollars)", y="Fast-Food Chains", hue="Fast-Food Chains", palette="magma", legend=False, dodge=False)
plt.title("Top 10 Chains by Average Sales per Unit")
plt.xlabel("Average Sales (in Thousands USD)")
plt.ylabel("Fast Food Chain")
plt.tight_layout()
plt.show()

# Correlation Heatmap of Numeric Features
plt.figure(figsize=(10, 6))
corr_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation heat map of numerical columns")
plt.tight_layout()
plt.show()

# Franchise vs Company Owned Comparison (Top 10 by Total Units)
top_units = df.sort_values(by="2021 Total Units", ascending=False).head(10)
top_units_plot = top_units.set_index("Fast-Food Chains")[["Franchised Stores", "Company Stores"]]
top_units_plot.sort_index().plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set2')
plt.title("Franchise vs Company Owned Stores (Top 10 by Total Units)")
plt.ylabel("Number of Stores")
plt.xlabel("Fast Food Chain")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Top 10 Chains by Unit Growth from 2020
top_growth = df.sort_values("Total Change in Units from 2020", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_growth, x="Total Change in Units from 2020", y="Fast-Food Chains", hue="Fast-Food Chains", palette="crest", legend=False, dodge=False)
plt.title("Top 10 Fast Food Chains by Unit Growth (2020 to 2021)")
plt.xlabel("Unit Change")
plt.ylabel("Fast Food Chain")
plt.tight_layout()
plt.show()

# Summary Statistics
print("Summary Statistics:")
print(df.describe())
```
